```python
def read_file(fpath):
    f = open(fpath)

    for line in f:
        print(line.rstrip())

    f.close()
    # return None
```

```python
read_file('Day1.ipynb')
```

```python
def num_lines(fpath):
    line_count = 0
    f = open(fpath, encoding='utf-8')

    for line in f:
        line_count += 1
    
    f.close()
    return line_count
```

```python
def count_words(fpath):
    word_count = 0
    f = open(fpath, encoding='utf-8')

    for line in f:
        words = line.split()
        word_count += len(words)
    
    f.close()
    return word_count
```

```python
count_words('alice.txt')
```
